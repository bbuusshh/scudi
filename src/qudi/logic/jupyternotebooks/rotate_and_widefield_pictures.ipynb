```python
import time
import matplotlib.pyplot as plt
```

```python
# check for needed modules
try:
    ello_logic
except:
    print('ello logic not loaded')
try:
    camera_logic_oli
except:
    print('camera logic not loaded')
```

```python
angles = [1,2]
images = {}

camera_logic_oli.set_exposure_time(1)

for angle in angles:
    ello_logic.set_rotor_pos(angle)
    time.sleep(1) # ello needs time to rotate
    img =  camera_logic_oli.return_image()
    images[angle] = img
```

```python
for angle in angles:
    img = images[angle]
    plt.matshow(img)
```
